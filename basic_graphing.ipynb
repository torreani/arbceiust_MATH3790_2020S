#### Let us first import some key packages

```python
import numpy as np  #numerical computation package
import pandas as pd #dataframe
import matplotlib.pyplot as plt #basic plotting package
import seaborn as sns  #more advanced graphing

#The following directs see graphcal output to local display 
%matplotlib inline  
```

```python
x = np.arange(0, 1, 0.1)
print(x)
```

#### Basic Plotting

- We will use a module called `matplotlib` to plot some simple graphs.

- This module provides functions which are very similar to MATLAB plotting commands.

```python
import matplotlib.pyplot as plt
%matplotlib inline

y = x*2 + 5
plt.plot(x, y)
```

```python
# Plotting a sine curve

from numpy import pi, sin
plt.figure(figsize=(10,8)) #instantiate the figure method of plt
x = np.arange(0, 2*pi, 0.01)
y = sin(x)
plt.plot(x, y)
plt.show()
```

#### Plotting a histogram

- We can use the `hist()` function in `matplotlib` to plot a histogram

```python
# Generate some random data
data = np.random.randn(1000)

plt.hist(data)
plt.show()

# Computing histograms as matrices
```

#### Computing histograms as matrices

- The function `histogram()` in the `numpy` module will count frequencies into bins and returns them as arrays.

```python
a1, a2 = np.histogram(data)
```

```python
print("Frequencies:", a1)
print("At bin values:", a2)
```

#### Now let us get some external data from a URL and work with it

```python
import pandas as pd  #The pandas dataframe package is loaded with the name pd

#iris is a dataframe that we load the csv data read in from the URL into
iris = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/iris.csv", index_col=0)
```

```python
#Take a look at the data: the head method of the iris dataframe prints out 5 rows by default
iris.head()
```

#### Basic Graphing

```python
#Here is a most basic plot of one of the variables

plt.figure(figsize=(12,8))  #instantiate plt and call its method figure with parameters for figuresize
plt.plot(iris['Sepal.Length'])  #the 'Sepal.Length' column is plotted
plt.xlabel('ID')  #x-axis is labeled "ID"
plt.ylabel('Sepal Length') #y-axis "Sepal Length"
plt.show()
```

#### <font color= 'blue'> Exercise 1:  Now it is your turn: Graph of the column "Petal.Length" </font>

```python
plt.figure(figsize=(10,5))
plt.title('Graph of column Petal.Length')
plt.plot(iris['Petal.Length'])
plt.xlabel('ID')
plt.ylabel('Petal Length')
plt.show()
```

#### If you want to see many columns graphed in the same plot

```python
# Let us first grab the column names
column_names = [column for column in iris]  #list comprehension example
print(column_names)  #check out what we got
```

```python
# We are going to use a for loop to plot the numerical variable columns.  Here is an example for loop:
for x in range(0, 3):
    print("We're on time %d" % (x))
```

```python
#Here is a way to find the maximum and minimum values the numerical data in a dataframe
#why, for example, the min() called twice?
mini = iris[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']].min().min()
maxi = iris[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']].max().max()
print(mini, maxi)
```

```python
plt.figure(figsize=(16,10))
for i in range(0,4):
        plt.plot(iris[column_names[i]])
plt.ylim(mini,maxi)  #the plot limited to the range of the data - now you know why min and max were calculated
plt.legend(['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'])  #makes it easier to read
plt.show()
```

### Subplots: Now if we want to see many columns graphed at the same time

```python
plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.plot(iris['Sepal.Length'])
plt.xlabel('Id')
plt.ylabel('Sepal Length')

plt.subplot(2,2,2)
plt.plot(iris['Sepal.Width'],'b')
plt.xlabel('Id')
plt.ylabel('Sepal Width')

plt.subplot(2,2,3)
plt.plot(iris['Petal.Length'],'b')
plt.xlabel('Id')
plt.ylabel('Petal Length')

plt.subplot(2,2,4)
plt.plot(iris['Petal.Width'],'r')
plt.xlabel('Id')
plt.ylabel('Petal Width')

plt.show()
```

#### Slicing Data Frames

```python
#Here we slice the first 4 columns of iris into data frame
iris1 = iris.loc[:, :'Petal.Width']
iris1.tail()
```

```python
#Here we slice the first 100 rows of iris into another dataframe
iris_rows100 = iris.loc[0:99]
iris_rows100.tail()
```

```python
##Another common but powerful display of data using box-and-whisker plots
plt.figure(figsize=(12,8))
iris.boxplot(['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']) #this does the job the box plots of all columns
plt.show()
```

#### Here are a bit more advanced graphing:  Regression line with a confidence band

```python
sns.lmplot(data = iris, x = 'Sepal.Length', y = 'Petal.Length')
plt.show()
```

```python
# The same data but now broken into their "species"
sns.lmplot(data = iris, x = 'Sepal.Length', y = 'Petal.Length',hue="Species")
plt.show()
```
